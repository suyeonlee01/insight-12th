### 4-3) 데이터프레임을 합쳐주세요

#### 1. 형태가 동일한 추가 데이터를 기존 데이터프레임과 합친 df4를 만들어주세요

```python
# 추가 데이터

new_customer_id = ['n1', 'n2', 'n3']
new_total_km = [300, 410, 280]
new_car = ['coupe', 'coupe', 'sedan']
new_local = ['창원', '서울', '부산']
new_owner = ['엄상호', '문수빈', '이창민']
efficiency = [17, 20.5, 19.87]

new_data={'고객ID': new_customer_id,
      '주행거리': new_total_km,
      '차종': new_car,
      '지역': new_local,
      '차주': new_owner,
        '연비': efficiency
}

columns = ['고객ID', '주행거리', '차종', '지역', '차주', '연비']


new_df = pd.DataFrame(new_data, columns=columns)
```

```python
# concat 을 사용해서 df4 만들기
df4=pd.concat([df,new_df])
```

```python
# 데이터프레임 확인
df4
```

아래 그림이 나오면 됩니다.



#### 2. 형태가 다른 데이터를 합쳐 df5를 만들어주세요

```python
# 추가 데이터

new2_customer_id = ['n5', 'n6']
new2_total_km = [280, 290]
new2_car = ['sedan', 'coupe']
new2_owner = ['박현기', '전조은']
new2_efficiency = [17.76, 21]
color = ['red', 'green']


new_data2={'고객ID': new2_customer_id,
      '주행거리': new2_total_km,
      '차종': new2_car,
      '차주': new2_owner,
          '연비': new2_efficiency,
           '도색': color
}

columns = ['고객ID', '주행거리', '차종', '차주', '연비', '도색']


new2_df = pd.DataFrame(new_data2, columns=columns)
```

```python
# merge를 사용해서 df5 만들기, how=outer 설정

df5=pd.merge(df, new2_df, how ="outer")
```

```python
# 데이터 확인

df5
```
