### Z-score
 Z-score          .   3         Z-score  3   -3        .

 Z-score     .

```python
def find_outlier_z(data, threshold=3):
    mean = np.mean(data)
    std = np.std(data)
    
    z_score = [(y-mean)/std for y in data] #

    masks = np.where(np.abs(z_score)>threshold)
    
    return masks
```

'compression-ratio'  ,   .

```python
sns.boxplot(x=df["compression-ratio"])
plt.show()

print(f"find_outlier_ze()    : {find_outlier_z(df['compression-ratio'])}")
```

### Tukey Fences


Tukey Fences( )      ,   (IQR, Interquartile Range)  . IQR   50%  ,    25%  25%    .

(Q1  1.5 * IQR)   (Q3 + 1.5 * IQR)     . 1.5   3               .

 Tukey Fances     .

```python
def find_outlier_by_Tukey(data):
    q1, q3 = np.percentile(data,[25,75])
    iqr = q3-q1
    
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    
    print(f'lower_bound {lower_bound.round(3)}, upper_bound {upper_bound.round(3)} .\n')
    
    # np.where   
    mask = np.where((data>upper_bound) | (data<lower_bound)) 
    
    # np.where Dataframe.index    
    #mask = data[(data['feature'] > upper_bound) | (data['feature'] < lower_bound)].index
    return mask
```

'compression-ratio'  ,   .

```python
sns.boxplot(x=df["compression-ratio"])
plt.show()

print(f"find_outlier_by_Tukey()    : {find_outlier_by_Tukey(df['compression-ratio'])}")
```

```python
#    (Turkey Fence)
def find_outlier_by_Tukey(data):
    q1, q3 = np.percentile(data,[25,75])
    iqr = q3-q1
    
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    
    print(f'lower_bound {lower_bound.round(3)}, upper_bound {upper_bound.round(3)} .\n')
    
    # np.where   
    mask = np.where((data>upper_bound) | (data<lower_bound)) 
    
    # np.where Dataframe.index    
    #mask = data[(data['feature'] > upper_bound) | (data['feature'] < lower_bound)].index
    return mask

#     
print("    ")
sns.boxplot(x=df["compression-ratio"])
plt.show()

#  mask  
mask = find_outlier_by_Tukey(df['compression-ratio'])

#    df 
for i in range(len(mask)):
    df = df.drop(index=mask[i])
df = df.reset_index(drop=True)

#     
print("    ")
sns.boxplot(x=df["compression-ratio"])
plt.show()
```
